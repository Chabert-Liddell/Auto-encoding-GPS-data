```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
optimizer = "adam"
nb_epochs = 1000
nz = 2
monit2 = pd.read_csv("save/monitor_vae_std_10k_%d_%s_%d.csv" % (nz, optimizer, nb_epochs))
monit_cvae2 = pd.read_csv("save/monitor_cvae_std_10k_%d_%s_%d.csv" % (nz, optimizer, nb_epochs))
nz = 3
monit3 = pd.read_csv("save/monitor_vae_std_10k_%d_%s_%d.csv" % (nz, optimizer, nb_epochs))
monit_cvae3 = pd.read_csv("save/monitor_cvae_std_10k_%d_%s_%d.csv" % (nz, optimizer, nb_epochs))
nz = 4
monit4 = pd.read_csv("save/monitor_vae_std_10k_%d_%s_%d.csv" % (nz, optimizer, nb_epochs))
monit_cvae4 = pd.read_csv("save/monitor_cvae_std_10k_%d_%s_%d.csv" % (nz, optimizer, nb_epochs))
nz = 5
monit5 = pd.read_csv("save/monitor_vae_std_10k_%d_%s_%d.csv" % (nz, optimizer, nb_epochs))
monit_cvae5 = pd.read_csv("save/monitor_cvae_std_10k_%d_%s_%d.csv" % (nz, optimizer, nb_epochs))
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.plot(monit2.loss)
plt.plot(monit3.loss)
plt.plot(monit4.loss)
plt.plot(monit5.loss)
plt.grid()
plt.ylabel("- ELBO")
plt.title("VAE")
plt.ylim(5, 12)
plt.subplot(2,2,2)
plt.plot(monit_cvae2.loss)
plt.plot(monit_cvae3.loss)
plt.plot(monit_cvae4.loss)
plt.plot(monit_cvae5.loss)
plt.grid()
plt.legend("2345", title = "nz")
plt.title("CVAE")
plt.ylim(5, 12)


plt.subplot(2,2,3)
plt.plot(monit2.dkl)
plt.plot(monit3.dkl)
plt.plot(monit4.dkl)
plt.plot(monit5.dkl)
plt.grid()
plt.xlabel("Epoch")
plt.ylabel("DKL")
plt.ylim(0, 5)
plt.subplot(2,2,4)
plt.plot(monit_cvae2.dkl)
plt.plot(monit_cvae3.dkl)
plt.plot(monit_cvae4.dkl)
plt.plot(monit_cvae5.dkl)
plt.grid()
plt.xlabel("Epoch")

plt.ylim(0, 5)
```
